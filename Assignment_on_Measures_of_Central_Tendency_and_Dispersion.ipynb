4. What do you think about the scoring variation for math, reading and writing
based on
* Gender
* Test preparation course

### Gender

```python
gender_var = data.groupby('gender').std()[['math score', 'reading score', 'writing score']]
gender_var
```

```python
# Plotting the scoring variation based on gender
gender_var.plot(kind='bar', figsize=(10, 6))
plt.title('Scoring variation by Gender')
plt.xlabel('Gender')
plt.ylabel('Scoring variation')
plt.xticks(rotation=0)
plt.legend(title='Scores')
```

* Overall, the standard deviations for reading and writing scores are similar between genders, indicating comparable levels of variation in performance.
* However, for math scores, females exhibit slightly higher variability compared to males.

### Test Preparation Course

```python
var_prep_course = data.groupby('test preparation course').std()[['math score', 'reading score', 'writing score']]
var_prep_course
```

```python
# Plotting the scoring variation based on Test Preparation Course
var_prep_course.plot(kind='bar', figsize=(10, 6))
plt.title('Scoring variation by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Scoring variation')
plt.xticks(rotation=0)
plt.legend(title='Scores')
```

* The standard deviation of math scores is lower for students who completed the test preparation course compared to those who did not.
* Similarly, the standard deviation of reading scores is lower for students who completed the test preparation course compared to those who did not.
* The standard deviation of writing scores is also lower for students who completed the test preparation course compared to those who did not.
* Overall, completing the test preparation course appears to lead to lower variability in scores across all subjects (math, reading, writing).
