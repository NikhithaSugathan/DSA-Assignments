4. Fill in all the missing values present in all the columns in the dataset

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
titanic_df.drop('Cabin',axis=1,inplace=True)
```

5. Check and handle outliers in at least 3 columns in the dataset

```python
titanic_df.columns
```

```python
num_col=['Pclass','Age','SibSp','Parch','Fare']
for i in num_col:
  plt.figure()
  plt.boxplot(titanic_df[i])
  plt.title(i)
```

```python
columns = ['Age','SibSp','Fare']

def handle_outliers_iqr(data, columns):
  for x in columns:
    Q1 = data[x].quantile(0.25)
    Q3 = data[x].quantile(0.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q3 + 1.5 * IQR

    # clipping
    data[x] = data[x].clip(lower=lower_lim, upper=upper_lim)
    outliers = data[(data[x] < lower_lim) | (data[x] > upper_lim)]

  return data


data = handle_outliers_iqr(titanic_df,columns)
data
```

```python
num_col = ['Age', 'SibSp','Fare']

for i in num_col:
  plt.boxplot(data[i])
  plt.title(i)
  plt.show()
```
